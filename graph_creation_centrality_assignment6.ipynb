```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import warnings
from datetime import datetime
from collections import Counter
%matplotlib inline
```

```python
#create graph
G=nx.Graph()
G.add_nodes_from([1,2,3,4,5,6])

print(nx.info(G))
```

```python
#add edges
G.add_edges_from([(1,2),(1,3),(1,6)])
G.add_edges_from([(2,6)])
G.add_edges_from([(3,4)])
G.add_edges_from([(4,6),(4,5)])
G.add_edges_from([(5,6)])
```

```python
nx.draw_networkx(G,node_color='r', node_size=405, with_labels=True)
```

```python
#reposition nodes
pos=nx.circular_layout(G)
nx.draw_networkx(G,node_color='r', node_size=405, with_labels=True)
```

```python


#get degree centrality
print('Degree centrality:','\n', nx.degree_centrality(G),'\n')

#get betweenness centrality
print('Betweenness centrality:','\n',nx.betweenness_centrality(G),'\n')

#get closeness centrality
print('Closeness centrality:','\n',nx.closeness_centrality(G))
```
